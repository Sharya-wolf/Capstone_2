# Core Marketing Analysis

Q1: Who are the territory managers for the sales territories assigned? What are the store IDs and cities for the stores in each assigned sales territory?

The territory managers for my sales territories assigned are Ellen Lemon & Lana llana.
The store IDs are listed from 865 down to 729. There are 19 different cities for the stores assigned to the sales territory together. Connecticut with 8 and Florida with 11.

```python
Territory_managers = StoreDetail[
    StoreDetail["State"].isin(["Florida", "Connecticut"])
    ]
Territory_managers.head()
```

Q2: What is the monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data?

```python
Sales = pd.merge(
    StoreSales,
    Territory_managers,
    on="Store ID",
    how= "inner"
) 
 # This only merges the two stores, which only gives us the first step of 
# answering the question.

Sales["Transaction Date"] = pd.to_datetime(Sales["Transaction Date"])
# Converting a datetime column

Sales["Month"] = Sales["Transaction Date"].dt.to_period("M")
# Makes a month column

Sales["Rank"] = Sales.groupby(
    "State")["Sale Amount"].rank(
    ascending=False)

monthly_revenue = Sales.groupby(
    ["State", "Month"])["Sale Amount"].sum().reset_index()
# By using .reset_index, it converts the series back into a DataFrame
print(monthly_revenue)
```

Q3: How would you rank the sales performance of each store in each sales territory? Which are the top-performing stores?

```python
store_performance = Sales.groupby(
    ["State", "Store ID", "Store Location"])["Sale Amount"].sum().reset_index()
# By doing this, we know the total sales for each store.

store_performance["Rank"] = store_performance.groupby(
    "State")["Sale Amount"].rank(
    ascending=False) # ranks stores within each territory

store_performance = store_performance.sort_values(
    ["State", "Rank"]
)
# This sorts the results of the ranking 

print(store_performance)
# Shows each store total revenue ranking within the territories Connecticut and Florida.
# As we can see for Connecticut, the top store is Store 870 in New London with 
# 331,116.76.
# Florida top ranking store is store 724 in Miami with 618,846.25.
```

## Visualization Revenue

```python
pivot_table = store_performance.pivot_table(
    index="Store Location",
    columns="State",
    values="Sale Amount",
    aggfunc="sum"
)

ax = pivot_table.plot(
    kind="bar",
    figsize=(14, 7)
)

plt.title("Store Revenue by Territory")
plt.xlabel("Store Location")
plt.ylabel("Revenue ($)")

plt.xticks(rotation=45, ha='right')

ax.yaxis.set_major_formatter(
    StrMethodFormatter('${x:,.0f}')
)

plt.tight_layout()

plt.show()
```

Q4: Comparing the customer ID from the customer list data with the rewards ID from the sales data, who were the top customers in each sales territory?

```python
customer_sales = pd.merge(
    customer_list, Sales,
    left_on="cust_id",
    right_on="RewardsID",
    how="inner") # This merges the tables and matches the keys, keeping only the matching customers.

top_customers = customer_sales.groupby(
    ["State", "cust_id", "name"])["Sale Amount"].sum().reset_index() 
# Here we are grouping the table by the territory and customers. using this we can 
# calculate the total spendings of each customer. The reason we don't group by store
# location is because it will mess with the ranking system. A customer could be shown 
# twice because they go to different stores. 

top_customers["Rank"] = top_customers.groupby(
    "State")["Sale Amount"].rank(
    ascending=False, method="dense")

top_customers[top_customers["Rank"] == 1]
# This is where the ranking of the customers comes in.

top_customers = top_customers.sort_values(
    ["State", "Rank"]
) # Always make sure to sort the results; it makes the ranking clear. It will show the
# best performers first.
top_customers
# The top customer in the sales territory of Connecticut is Marie S. with $4,338.87.
# The top customer in the sales territory of Florida is Harry Morgan with $19.87.
```

Q5: What is the number of transactions per month by product category in each assigned territory? What is the total sales revenue per month by category? What might this tell you about the most popular products, and where could there be opportunity for growth?

```python
Sales_products = pd.merge(
    Sales,
    Products,
    on="Prod Num")

Sales_category = pd.merge(
    Sales_products,
    ProductCategories,
    on="CategoryID")
# Since Sales doesn't have anything to match with Category, we have to merge twice. First
# with Products to get the CategoryID for us to merge with Category to get the Categories.

Sales_category["Transaction Date"] = pd.to_datetime(
    Sales_category["Transaction Date"])
# Since the question is asking for the month, we must convert the column of dates first.

Sales_category["Month"] = Sales_category["Transaction Date"].dt.to_period("M")
# This creates the month column the question wants per month specifically. dt refrencing
# the datetime and ("M") representing month.

transactions_per_month = Sales_category.groupby(
    ["State", "Month", "Category"])["CategoryID"].count().reset_index()
# We use the count function so we can answer the question of how many transactions per
# month.
transactions_per_month = transactions_per_month.rename( 
    columns={"CategoryID": "Transaction Count"})
# This answers the number of transactions per month

transactions_per_month = transactions_per_month.sort_values(
    ["State", "Month", "Category"])
# This sorts the results
transactions_per_month
```

```python
monthly_category_revenue = Sales_category.groupby(
    ["State", "Month", "Category"])["Sale Amount"].sum().reset_index()
monthly_category_revenue
```

```python
pivot_table = monthly_category_revenue.pivot_table(
    index="Category",
    columns="State",
    values="Sale Amount",
    aggfunc="sum"
)

ax = pivot_table.plot(
    kind="bar",
    figsize=(12, 6)
)

plt.title("Monthly Sales Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Sale Amount ($)")
plt.xticks(rotation=30, ha='right')

ax.yaxis.set_major_formatter(
    StrMethodFormatter('${x:,.0f}')
)

plt.tight_layout()
plt.show()
```
